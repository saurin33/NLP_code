```python
import pandas as pd 
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('fake_or_real_news.csv')
df
```

```python
X = df['text']
y = df.label
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
count_vectorizer = CountVectorizer(stop_words = 'english')
```

```python
count_train = count_vectorizer.fit_transform(X_train, y_train)
```

```python
count_test = count_vectorizer.transform(X_test)
```

```python
print(count_vectorizer.get_feature_names()[:10])
```

```python
count_df = pd.DataFrame(count_train.A, columns = count_vectorizer.get_feature_names())
count_df
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_df = 0.7)
```

```python
tfidf_train = tfidf_vectorizer.fit_transform(X_train, y_train)
```

```python
tfidf_test = tfidf_vectorizer.transform(X_test)
```

```python
print(tfidf_vectorizer.get_feature_names()[:10])
```

```python
print(tfidf_train.A[:5])
```

```python
tfidf_df = pd.DataFrame(tfidf_train.A, columns = tfidf_vectorizer.get_feature_names())
tfidf_df
```

```python
#  Calculate the difference in columns 

difference = set(count_df.columns) - set(tfidf_df.columns)
print(difference)
```

```python
# Check if the columns are equal

print(count_df.equals(tfidf_df))
```

```python
nb_classifier = MultinomialNB()
```

```python
nb_classifier.fit(count_train, y_train)
```

```python
pred = nb_classifier.predict(count_test)
```

```python
score = metrics.accuracy_score(y_test, pred)
print(f'Accuracy Score: {score}')
```

```python
cm = metrics.confusion_matrix(y_test, pred, labels=['FAKE', 'REAL'])
print(cm)
```

```python
print("Classification Report: \n {}"  .format(classification_report(y_test, pred, labels=["FAKE", "REAL"])))
```

```python
#  Tfidf model
```

```python
nb_classifier.fit(tfidf_train, y_train)
```

```python
pred_y = nb_classifier.predict(tfidf_test)
```

```python
print(f'Accuracy Score: {metrics.accuracy_score(y_test, pred_y)}')
```

```python
print(f'Confusion Matrix: \n {metrics.confusion_matrix(y_test, pred_y, labels=["FAKE", "REAL"])}')
```

```python
print("Classification Report: \n {}".format(classification_report(y_test, pred_y, labels=["FAKE", "REAL"])))
```

```python
#  Use different alpha 

alphas = np.arange(0, 1, 0.1)

def train_and_predict(alpha):
    nb_classifier.fit(tfidf_train, y_train)
    y_pred = nb_classifier.predict(tfidf_test)
    score = metrics.accuracy_score(y_test, y_pred)
    return score

for alpha in alphas:
    print('Alpha: ', alpha)
    print('Accuracy Score: ',  train_and_predict(alpha))
    print()
```

```python
class_labels = nb_classifier.classes_
```

```python
feature_names = tfidf_vectorizer.get_feature_names()
```

```python
feat_with_weights = sorted(zip(nb_classifier.coef_[0], feature_names))
```

```python
# First 20 & last 20 class labels

print(class_labels[0], feat_with_weights[:20])
print(class_labels[1], feat_with_weights[-20:])
```
